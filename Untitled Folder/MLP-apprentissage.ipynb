# L'apprentissage d'un réseau de neurones

## 0) Objectif

Montrer **comment** on fait apprendre un réseau de neurones
Le problème à résoudre est celui des 2 lunes imbriquées 
C'est un problème classique de la Data Science.

Le réseau doit apprendre à reconnaitre les points de chaque lune durant la phase d'apprentissage.

Ensuite, en production, le réseau pourra distinguer chaque nouveau point non vu.

### chargement des librairies

```
# scikit learn : librairie de Machine Learning
from sklearn import datasets
from sklearn import model_selection

# keras : librairie de Deep Learning
from keras.models import Sequential
from keras.layers import Dense

# numpy : utilitaire de calculs
import numpy as np

# matplotlib : librairie graphique
import matplotlib.pyplot as plt
```

### figer le générateur de nombres aléatoires

```
# fix random seed 
seed = 12345
np.random.seed(seed)
```

## 1) Les Datas
Générer un data set : 2 lunes imbriquées l'une dans l'*autre*

```
X,y = datasets.make_moons(n_samples = 1500, shuffle = True, noise = 0.3, random_state = seed)
```

###  Un data set synthétique

On génère un data set synthétique grâce à la librairie [Scikit-Learn](https://scikit-learn.org/stable/). 

### Pourquoi un data set synthétique ?

C'est très utile lorsqu'on débute ou lorsqu'on met au point un nouveau modèle de pouvoir contrôler la difficulté de la tâche qu'on demande au réseau.

Paramètres : 
- "noise" est le niveau de bruit qu'on injecte dans les datas
- n_samples : nb de points générés

#### description du data set
2 lunes imbriquées. 

Les données sont étiquettées "0" ou "1" selon qu'elles appartiennent à l'une ou l'autre des 2 lunes.

```
#Affichage des étiquettes des premiers points du nuage
print (y[0:5])
```

```
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33, random_state= seed)
```

## 2) Le graphique
Faire un graphique en nuage des 2 lunes imbriquées

```
plt.figure()
plt.scatter(X[: , 0] , X[: , 1] , c=y)
```

## 3) Créer le réseau de neurones

Marche à suivre :
On utilise l'API sequential de Keras.

Ensuite, on crée les couches (entrée - cachée - sortie)

On compile le modèle



```
model = Sequential()
# taille du réseau : 1 neurone 
nb_neurone = 1
# input_dim = 2 (X et y). Fonction activation : sigmoïde
model.add(Dense(nb_neurone , input_dim=2,  activation='sigmoid'))
model.add(Dense(nb_neurone , activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Le réseau est prêt. On demande affichage de la structure du réseau

```
model.summary()
```

## 4) La phase d'Apprentissage

Durant la phase d'apprentissage ("*fit*"), on montre les exemples au réseau pour qu'il apprenne. 

En Deep Learning, l'objectif de l'apprentissage est d'apprendre à réaliser un mapping complexe entre les données d'entrée et les sorties.

Les inputs sont les coordonnées des points représentant les 2 lunes imbriquées.

Les outputs sont les classes, les labels ("0" ou "1") distinguant les 2 lunes

### Batch size
Durant l'apprentissage, on groupe les données d'entrée (*X*). On réalise des groupes comprenant "*batch_size*" points. Dans notre cas, *batch_size* = 10. Donc, 10 données d'entrée par groupes.

La constitution des groupes est faite au hasard, on mélange les points.

### algorithme d'apprentissage 
* on présente un groupe (*batch*) de données d'entrée (*input*). Ici ce sont des batch de 10 points
* pour chacune donnée de ce groupe, le réseau réalise une prédiction de la classe à laquelle appartient ce point.
* on compare la prédiction avec la vraie valeur (la classe à laquelle appartient réellement le point)

Bien entendu, on ne montre pas au réseau cette vraie valeur. C'est ce qu'il doit apprendre ;-) 
* on calcule la valeur de la fonction d'erreur ("*binary_crossentropy*")
* si on a passé toutes les données d'entrée du *batch* en cours alors
on modifie les connections du réseau afin de diminuer l'erreur au prochain tour 
* on change de batch
* on fait l'ensemble de ces opérations **epoch** fois 
 (dans notre cas, *epoch* = 50)

### Méthode *fit* de Keras 
Ce sont toutes ces opérations qui seront à chaque fois durant l'apprentissage.
Grace à l'API de haut niveau Keras, ceci est transparent.
En effet, pour réaliser tous les points que nous venons de voir, il suffit juste d'appeller la méthode **fit**.

Le *verbose* fixe la sortie, le report selon le niveau de détail que l'on souhaite.

*verbose* = 0 pour **pas** de sortie du tout. Sinon, *verbose* peut prendre les valeur de 1 à 4




```
#On garde 1/3 des exemples pour la phase de test (*validation_data*)
model.fit(X, y, validation_data =[X_test  , y_test], epochs=50, batch_size=10, verbose=2)
```

## 5)  Report : 

Durant la phase d'apprentissage, Keras nous indique quel est le taux d'erreur sur les données non vues. 

Ce taux d'erreur figure dans le report : *val_acc :*

Ces données non-vues présentées durant la phase d'apprentissage servent **uniquement** a calculer le pourcentage de bonnes prédictions.

On peut recalculer ce taux d'erreur. Code suivant :

```
score, acc = model.evaluate(X_test , y_test)

# accuracy : % instances bien classées
print("% bonnes prédictions = {:.2f}").format(100 * acc)
```

## 6) Phase de test 

on génère des prédictions sur les données non vues
On présente au réseau les données laissées de côté durant l'apprentissage (*X_test*).

On demande au réseau entrainé de produire une prédiction pour chacun de ces points
*predict_classes*

```
predictions = model.predict_classes(X_test)
# redimension du tableau contenant les predictions du réseau
pred =  predictions.reshape(-1,)
```

```
plt.figure()
plt.scatter(X[: , 0] , X[: , 1], c=y)

# indice (position dans le tableau) des points mal classés par le réseau (prédictions fausses)
idx = np.where(y_test != pred)

# les points mal classés (en rouge)
plt.scatter(X_test[idx , 0] , X_test[idx , 1] , c = 'red')
```
