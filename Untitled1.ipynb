```python
import nltk
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

import pandas as pd
import string
import seaborn as sns
```

```python
df1=pd.read_excel('final_210504_tous labels.xlsx',engine='openpyxl')
```

```python
df1["Messages"]=df1["Messages"].astype(str)
```

```python
df1["Length"] = df1["Messages"].apply(len)
```

```python
df=df1.head(1900)
```

```python
class PreProcessText(object):
    def __init__(self):
        pass
    
    def __remove_punctuation(self, text):
        """
        Takes a String 
        return : Return a String 
        """
        message = []
        for x in text:
            if x in string.punctuation:
                pass
            else:
                message.append(x)
        message = ''.join(message)
        
        return message
    
    def __remove_stopwords(self, text):
        """
        Takes a String
        return List
        """
        words= []
        for x in text.split():
            if x.lower() in stopwords.words('french'):
                pass
            else:
                words.append(x)
        return words
    
    
    def token_words(self,text=''):
        """
        Takes String
        Return Token also called  list of words that is used to 
        Train the Model 
        """
        message = self.__remove_punctuation(text)
        words = self.__remove_stopwords(message)
        return words
```

```python
mess = 'exemple de message! Notice: ceci est une ponctuation.'
obj = PreProcessText()
words = obj.token_words(mess)
print(words)
```

```python
bow_transformer = CountVectorizer(analyzer=obj.token_words).fit(df["Messages"])
```

```python
messages_bow = bow_transformer.transform(df["Messages"])
```

```python
tfidf_transformer = TfidfTransformer().fit(messages_bow)
```

```python
messages_tfidf = tfidf_transformer.transform(messages_bow)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB().fit(messages_tfidf,df["catégorie"])
```

```python
all_predictions = model.predict(messages_tfidf)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix

print('accuracy %s' % accuracy_score(all_predictions, messages_tfidf))
print(classification_report(messages_tfidf, all_predictions,labels=all_predictions))
```
